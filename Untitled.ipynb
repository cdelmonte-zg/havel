```python
from jsonschema2db import JSONSchemaToPostgres
import json
import psycopg2
schema = json.load(open('./kdsf-basis.json'))
translator = JSONSchemaToPostgres(
    schema,
    postgres_schema='schm'
)

pg_connection_dict = {
    'dbname': 'postgres',
    'user': 'postgres',
    'password': 'changeme',
    'port': '5432',
    'host': 'localhost'
}

print(pg_connection_dict)
con = psycopg2.connect(**pg_connection_dict)
translator.create_tables(con)

```
